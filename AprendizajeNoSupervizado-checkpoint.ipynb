```python
numeric_columns = data.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = data.select_dtypes(include=['object']).columns

scaler = StandardScaler()
scaled_numeric_data = pd.DataFrame(scaler.fit_transform(data[numeric_columns]), 
                                  columns=numeric_columns)
if len(categorical_columns) > 0:
    categorical_data = pd.get_dummies(data[categorical_columns])
    X = pd.concat([scaled_numeric_data, categorical_data], axis=1)
else:
    X = scaled_numeric_data

X
```

## K-Means

```python
Nc = range(1, 9)
kmeans = [KMeans(n_clusters=i) for i in Nc]

score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]

plt.plot(Nc, score, marker='o')
plt.xlabel('Numero de Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans = KMeans (n_clusters=3).fit(X) 
centroids = kmeans.cluster_centers_ 
centroids
```

```python
# Obtenemos las etiquetas de cada punto de nuestros datos
# Obtenemos las etiquetas de cada punto
labels = kmeans.labels_

# Obtenemos los centroides
C = kmeans.cluster_centers_

# Asignamos colores
colores = ['red', 'green', 'cyan']
asignar = [colores[i] for i in labels]
```

```python
# Usamos columnas reales del dataset
f1 = X['total_bill'].values
f2 = X['tip'].values

# Graficamos los puntos y los centroides
plt.scatter(f1, f2, c=asignar, s=70)
plt.scatter(C[:, 0], C[:, 1], marker='*', c=colores, s=1000)
plt.title('Clusters sobre total_bill y tip')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
f1 = X['size'].values
f2 = X['tip'].values

plt.scatter(f1, f2, c=asignar, s=70)
plt.scatter(C[:, 0], C[:, 1], marker='*', c=colores, s=1000)
plt.title('Clusters sobre size y tip')
plt.xlabel('size')
plt.ylabel('tip')
plt.show()
```

## Hierarchical clustering

```python
ward_clustering = linkage(X, method="ward", metric="euclidean")
# Graficar el dendrograma
plt.figure(figsize=(20, 7))
dendrogram(ward_clustering)
# Línea horizontal para sugerir número de clústeres
plt.axhline(y=10, color='r', linestyle='--')
# Etiquetas
plt.title("Dendrograma - Clustering Jerárquico")
plt.xlabel("Índice de muestra")
plt.ylabel("Distancia Euclidiana")
plt.show()
```

```python
hier_clustering = AgglomerativeClustering(n_clusters=3)
labels_hier = hier_clustering.fit_predict(X)
labels_hier
```

```python
# Obtenemos las etiquetas de cada punto de nuestros datos
labels = labels_hier

colores = ['red', 'green', 'cyan']
asignar = []

for row in labels:
    asignar.append(colores[row])
```

```python
f1 = X['total_bill'].values  
f2 = X['tip'].values 
plt.scatter(f1, f2, c=asignar, s=70) 
plt.title('Datos etiquetados')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
# Calcular medidos de calidad para Clustering Jerárquico

print("\nMedidas de calidad рага Clustering Jerárquico:")

print("Puntaje de silueta:", metrics.silhouette_score(X, labels_hier)) 

print("Indice de Calinski-Harabasz:", metrics.calinski_harabasz_score(X, labels_hier))
print("Davies-Bouldin Index:", metrics.davies_bouldin_score(X, labels_hier))
```

```python
# Calcular el coeficiente de correlación cofenética
from scipy.cluster.hierarchy import linkage

# Compute the distances

X_dist = pdist(X)

# Compute the Linkage matrix

Z = linkage(X, "ward")

# Calculate the cophenetic correlation coefficient
coef_cofenetico = cophenet(Z, X_dist)

print("Coeficiente de correlación cofenética:", coef_cofenetico)
```
